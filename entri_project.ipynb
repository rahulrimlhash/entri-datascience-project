```python
# Here I'm importing all the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the dataset
data = pd.read_csv('data.csv')
```

```python
# data preprocessing
```

```python
data['Height'] = np.random.randint(150, 181, size=len(data))
```

```python
data.head()
```

```python
# Fill missing values in the "salary" column with the mean value
mean_salary = data['Salary'].mean()
data['Salary'].fillna(mean_salary, inplace=True)
```

```python
data.head()
```

```python
# Fill missing values in the "college" column with the mode value
mode_college = data['College'].mode().iloc[0]
data['College'].fillna(mode_college, inplace=True)
```

```python
data.head()
```

```python
# Droping unwanted columns
data = data.drop('Number', axis=1)
```

```python
data
```

#### How many are there in each Team and the percentage splitting with respect to the total employees.


```python
# Count employees in each team
team_counts = data['Team'].value_counts()

# Calculate percentage splitting
total_employees = len(data)
team_percentages = (team_counts / total_employees) * 100

# Display team counts and percentages
print("Team Counts:")
print(team_counts)
print("Team Percentages:")
print(team_percentages)
```

#### Segregate the employees with respect to different positions.


```python
# Segregate employees by position
position_counts = data.groupby('Position').size()

# Display position counts
print("Position Counts:")
print(position_counts)
```

```python
dataset = data
```

#### Find from which age group most of the employees belong to.



```python
# Categorize age into groups
age_groups = pd.cut(data['Age'], bins=[20, 30, 40, 50, 60, 70])

# Find the most common age group
most_common_age_group = age_groups.value_counts().idxmax()

# Display the most common age group
print("Most Common Age Group:")
print(most_common_age_group)
```

#### Find out under which team and position, spending in terms of salary is high.


```python
# Calculate total salary spending for each team and position
salary_spending = data.groupby(['Team', 'Position'])['Salary'].sum()

# Find the team and position with high salary spending
high_salary = salary_spending.idxmax()

# Display the team and position with high salary spending
print("Team and Position with High Salary Spending:")
print(high_salary)
```

#### Find if there is any correlation between age and salary , represent it visually.


```python
# Calculating the correlation between age and salary
correlation = data['Age'].corr(dataset['Salary'])

# Creating a scatter plot of age vs salary
plt.scatter(dataset['Age'], dataset['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary')
plt.show()

# Display the correlation coefficient
print("Correlation Coefficient:")
print(correlation)
```
