```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("black_friday_sale.csv")
```

```python
# Remove the columns "Product_Category_2" and "Product_Category_3"
data.drop(columns=["Product_Category_2", "Product_Category_3"], inplace=True)
```

```python
data.isnull().sum()
```

```python
# Convert the "Age" column to numeric representation
data["Age"] = data["Age"].map({
    '0-17': 1,
    '18-25': 2,
    '26-35': 3,
    '36-45': 4,
    '46-50': 5,
    '51-55': 6,
    '55+': 7
})

# Prepare the data for K-means clustering
X = data[["Age", "Purchase"]]

# Perform K-means clustering with 5 clusters (you can adjust the number of clusters as needed)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data["Cluster"] = kmeans.fit_predict(X)
```

```python
data.head()
```

```python
# Find the buyer's age group and their product of interest
age_group_product_interest = data.groupby("Age")["Product_ID"].apply(lambda x: x.value_counts().idxmax())
```

```python
age_group_product_interest
```

```python
# Identify the product category with the highest sales
highest_sales_product_category = data.groupby("Product_Category_1")["Purchase"].sum().idxmax()
```

```python
highest_sales_product_category
```

```python
print("Buyer's Age Group and their Product of Interest:")
print(age_group_product_interest)

print("\nProduct Category with the Highest Sales:")
print(highest_sales_product_category)
```
